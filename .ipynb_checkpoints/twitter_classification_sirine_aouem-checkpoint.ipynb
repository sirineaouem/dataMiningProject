<b style="font-size:18px;color:#43ff79;">Pretraitement des donn√©es:</p>

<p>dans cette partie on va utilis√© RegEx</p>

<p>Un RegEx, ou expression r√©guli√®re, est une s√©quence de caract√®res qui forme un mod√®le de recherche.

Le RegEx peut √™tre utilis√© pour v√©rifier si une cha√Æne contient le mod√®le de recherche sp√©cifi√©.</p>

<p>Python poss√®de un paquet int√©gr√© appel√© re, qui peut √™tre utilis√© pour travailler avec des expressions r√©guli√®res.</p>

<p> Dans re.sub(), on sp√©cifie un mod√®le d'expression r√©guli√®re dans le premier argument, une nouvelle cha√Æne dans le deuxi√®me argument, et une cha√Æne √† traiter dans le troisi√®me argument</p>

<p>Nous pouvons combiner un mod√®le d'expression r√©guli√®re en objets de type "pattern", qui peuvent √™tre utilis√©s pour la comparaison de mod√®les. Il est √©galement possible de rechercher √† nouveau un mod√®le sans le r√©√©crire grace la fonction re.compile</p>

<p>maintenant on commence √† nettoyer les donn√©es en √©liminant les donees unitiles </p>

```python
import re
for index, row in twitter_data.iterrows():
    text= row['text']
    httpdata= re.sub(r"http\S+", "", text)
    
    hashdata = re.sub(r"#\S+", "", httpdata)
    atdata = re.sub(r"@\S+", "", hashdata)
    brdata = re.sub(r"\n+", "", atdata)
    rtdata = re.sub(r"RT+", "", brdata)
    numddata = re.sub('[0-9]+', "", rtdata)
    laughdata = re.sub("hhh+", '', rtdata)
    emoji = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    newdata = re.sub(emoji, "", laughdata)
    twitter_data.loc[index,'text'] =newdata
```

<p>La premi√®re manipulation souvent effectu√©e dans le traitement du texte est la suppression de ce qu'on appelle en anglais les stopwords. Ce sont les mots tr√®s courants dans la langue √©tudi√©e ("and", "a", "the"... ) qui n'apportent pas de valeur informative pour la compr√©hension du "sens" d'un document et corpus. Il sont tr√®s fr√©quents et ralentissent notre travail : nous souhaitons donc les supprimer.</p>

<p>Il existe dans la librairie NLTK une liste par d√©faut des stopwords dans plusieurs langues, nous allons faire ceci d'une autre mani√®re √† supprimer les mots les plus fr√©quents du corpus et consid√©rer qu'il font partie du vocabulaire commun et n'apportent aucune information.</p>

<p>Plus qu'une derni√®re √©tape et on en aura termin√© avec le pr√©traitement !

Le processus de ¬´ lemmatisation ¬ª consiste √† repr√©senter les mots (ou ¬´ lemmes ¬ª) sous leur forme canonique. Par exemple pour un verbe, ce sera son infinitif. Pour un nom, son masculin singulier. L'id√©e √©tant encore une fois de ne conserver que le sens des mots utilis√©s dans le corpus.

Si l'on reprend notre exemple pr√©c√©dent, "Bonjour, je suis un texte d'exemple pour le cours d'Openclassrooms. Soyez attentifs √† ce cours !"

L'id√©al serait d'extraire les lemmes suivants : ¬´ bonjour, √™tre, texte, exemple, cours, openclassrooms, √™tre, attentif, cours ¬ª. Dans le processus de lemmatisation, on transforme donc ¬´ suis ¬ª en ¬´ √™tre¬ª  et ¬´ attentifs ¬ª en ¬´ attentif ¬ª.</p>

```python
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
Porter_steamer = PorterStemmer() 
vectorizer = CountVectorizer()
stem_data=[]
for i in range(0,twitter_data.shape[0]):
    stemmed_array=twitter_data['text'][i].split()
    stemmed=[Porter_steamer.stem(word) for word in stemmed_array if not word in set(stopwords.words('english'))]
    stemmed=' '.join(stemmed)
    stem_data.append(stemmed)
print(stem_data[0:20])
```

<b style="font-size:20px;color:#ff4d4d;"> Partie III:Classification des tweets</b>

<p>tant donn√© un ensemble de tweets, l‚Äôobjectif est de les r√©sumer sous formes de groupes de sorte √† ce que les Tweets qui sont dans le m√™me groupe soient similaires. Ainsi, l‚Äôutilisateur pourra par la suite lire juste un Tweet de chaque groupe (le Tweet qui est le centro√Øde de groupes).</p>

<p>pour cela on va utiliser la classification k-means: C‚Äôest l‚Äôun des algorithmes de clustering les plus r√©pandus. Il permet d‚Äôanalyser un jeu de donn√©es caract√©ris√©es par un ensemble de descripteurs, afin de regrouper les donn√©es ‚Äúsimilaires‚Äù en groupes (ou clusters).

La similarit√© entre deux donn√©es peut √™tre inf√©r√©e gr√¢ce √† la ‚Äúdistance‚Äù s√©parant leurs descripteurs ; ainsi deux donn√©es tr√®s similaires sont deux donn√©es dont les descripteurs sont tr√®s proches. Cette d√©finition permet de formuler le probl√®me de partitionnement des donn√©es comme la recherche de K ‚Äúdonn√©es prototypes‚Äù, autour desquelles peuvent √™tre regroup√©es les autres donn√©es.</p>

<p>on a utiliser l‚Äôalgorithme K-Means pour  classer  les Tweets  en 30 classes et on va ajouter les vecteurs resultatnt de CountVectorizer commme donn√©es pour √™tres classifi√©es  </p>

```python
from sklearn.cluster import KMeans
dist=[]
for i in range(1,30):
    Kmeans=KMeans(n_clusters=i,
                  init='k-means++',#selects initial cluster centers for k-mean clustering in a smart way to speed up convergence.
                  random_state=0,#Determines random number generation for centroid initialization. Use an int to make the randomness deterministic
                  verbose=True)
    Kmeans.fit(matrix)
    dist.append(Kmeans.inertia_)#Sum of squared distances of samples to their closest cluster center.
```

<p> Pour chaque it√©ration un score a √©t√© affich√©</p>

<p>La m√©thode Elbow est une technique tr√®s populaire et l'id√©e est d'effectuer le k-means clustering  pour un nombre de clusters k  et pour chaque valeur, nous calculons la somme des carr√©s des distances de chaque point √† son centre assign√© (distorsions).

Lorsque les distorsions sont trac√©es et que le graphique ressemble √† un bras, le "elbow" (le point d'inflexion de la courbe) est la meilleure valeur de k.</p>

```python
import matplotlib.pyplot as plt
plt.xlabel('num√©ro de clusters')
plt.ylabel('distance')
plt.plot(range(1,30),dist)
plt.title('methode elbow')
plt.show()
```

```python
true_k=30
Kmeans=KMeans(n_clusters=true_k,init='k-means++',n_init=1)
Kmeans.fit(matrix)
```

<p>maintenant on va afficher les mots racine essentiels pour chaque cluster comme un r√©sum√© de chaque groupeement </p>

```python
print("les mots essentielles du th√®me ") 
centroids = Kmeans.cluster_centers_.argsort()[:, ::-1]
mot = vectorizer.get_feature_names()
for i in range(true_k):
    print("dans Cluster %d:" % i)
    for ind in centroids[i, :10]:
        print(' %s' % mot[ind])
    print()
```

<p>Apr√®s  avoir  r√©cup√©r√©  les  classes,on a choisi un  Tweet  par  classe  comme  repr√©sentant.  Les  Tweets choisis seront les r√©sum√©s de toutes les informations contenues dans les Tweets.</p>

```python
final_twitter_data= pd.DataFrame(columns = [ 'text'])
i=0
j=0
while i<30:
    while True: 
        lastData=vectorizer.transform([stem_data[j]])
        prediction=Kmeans.predict(lastData)
        if i == prediction:
            print("represantant du cluster "+str(prediction)+twitter_data.loc[j,"text"])
            final_twitter_data.loc[i,"text"] =twitter_data.loc[j,"text"]
            j=0
            break
        j+=1
    i+=1
final_twitter_data.to_csv('final_twitter_dataFrame.csv',index = False )
```
